```python
import pandas as pd
import numpy as np
import time 

data = pd.read_csv("./datasets/done_data.csv", index_col=0)
data.describe()
data.head()
```

```python
from IPython.core.pylabtools import figsize
figsize(40, 7)

import matplotlib
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
for mm in data.tic.unique():
    y = data[data['tic'] == mm].adjcp
    x = range(len(y)) #data[data['tic'] == 'AAPL'].datadate
    plt.plot(x,y)
plt.show()
plt.close()
```

```python
import os, sys
dir = os.path.dirname('/Users/tinggao/git')
pathname = os.path.join(dir, '/usr/local/lib/python3.8/site-packages')
sys.path.append(os.path.realpath(pathname))

import gym
import seaborn as sns

for mm in ['adjcp', 'volume', 'macd', 'rsi', 'cci', 'adx']:  #'open', 'high', 'low',
    plt.figure(figsize=(8,5), dpi= 80)
    df = data[['tic', mm]]
    for tic in df.tic.unique():
        y = df[df['tic'] == tic][mm].values
        sns.kdeplot(y, alpha=.7)
    
    # Decoration
    plt.title('Density Plot on: ' + mm , fontsize=22)
    plt.legend()
    plt.show()
```

```python
plt.figure(figsize=(10,10), dpi= 80)

for mm in data.tic.unique():
    plt.scatter(data[data['tic'] == mm]['adjcp'], np.log(data[data['tic'] == mm]['volume']), label = mm, alpha=0.5 )

plt.legend()
plt.show()
```
